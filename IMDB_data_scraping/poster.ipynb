```python
import requests
import time
from bs4 import BeautifulSoup
import pandas as pd
import re
import os

# print(os.getcwd())

# url = 'https://www.imdb.com/search/title?count=100&release_date=2018,2018&title_type=feature'
# html = requests.get(url)
# # Parse html test
# html_soup = BeautifulSoup(html.text, 'html.parser')
# # Get the position of movies
# movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-advanced')
# img_url = movie_containers[1].find('img')['loadlate']

# name_link = movie_containers[1].find('a')['href']
# name = movie_containers[1].h3.a.text
# title_link = 'https://www.imdb.com' + str(name_link)
# html_link = requests.get(title_link)
# html_soup_link = BeautifulSoup(html_link.text, 'html.parser')
# # tag = html_soup_link.find('div', class_ = 'slate_wrapper')
# img_url = html_soup_link.find('img')['src']
# print(img_url)
    
        
def get_pic_from_url(url):
    # Using try except to avoid crawl stop
    try:
        # Get the url to local 
        html = requests.get(url)
        # Parse html test
        html_soup = BeautifulSoup(html.text, 'html.parser')
        # Get the position of movies
        movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-advanced')

        count=0
        for container in movie_containers:
            # The title and link of movie
            name = container.h3.a.text
            name_link = container.find('a')['href']
            title_link = 'https://www.imdb.com' + str(name_link)

            # Through bs to get the position of pictures
            html_link = requests.get(title_link)
            html_soup_link = BeautifulSoup(html_link.text, 'html.parser')
            tag = html_soup_link.find('div', class_ = 'poster')
            img_url = tag.find('img')['src']

            # Download the figure to local 
            with open('img/'+name+'.jpg','wb+') as f:
                f.write(requests.get(img_url).content)
            count = count+1
            print(count)
    except:
        print("There may be some problems")

def main(base_url):
    url_list = []
    # List all url should be crawled
    for i in range(2008,2019):
        url_list.append(base_url + str(i) + ',' + str(i) + '&title_type=feature')

    # Run functions and adding multiple df into one csv
    for url in url_list:
        df = get_pic_from_url(url)
    print('OK')
    
base_url = 'https://www.imdb.com/search/title?count=100&release_date='

if __name__ == '__main__':
    main(base_url)
```
